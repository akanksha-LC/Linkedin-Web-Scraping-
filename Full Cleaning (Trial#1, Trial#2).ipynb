```python
indx = -1
num = 0
for x in linkRaw1000['Industry']:
    indx += 1
    if isinstance(x, float):
        num+=1
    else:
        if(len(x.split(', '))>=2):
            x_list = x.split(', ')
            for state in states_list:
                if(x_list[1]==state):
                    linkRaw1000.Location[indx] = x
                    linkRaw1000.Industry[indx] = np.NaN
linkRaw1000['Location'][753] = 'Atlanta, Georgia'
linkRaw1000['Industry'][753] = np.NaN
```
