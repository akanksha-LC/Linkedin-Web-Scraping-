```python
# this is to make a new list with just the parameters I need, and not the slogans 
sloganlessList = []
chars = set('\n')
for x in rawList:
    if any((c in chars) for c in x):
        sloganlessList.append(x)
#print(newlisttry)
len(sloganlessList)
```

```python
#this is a new list to split the different parameters by "•"
periodLess = []
for x in sloganlessList:
    periodLess.append(x.split('•'))

#newlist0
```

```python
#this is a new list to split the parameters by "\n"
finalList = []
for x in periodLess:
    if(len(x)>1):
        finalList.append(x[0].split('\n') + x[1].split('\n'))
    else:
        finalList.append(x[0].split('\n'))

#newlist4
```

```python
#this is the raw data frame 
import pandas as pd
linkRaw1000 = pd.DataFrame(data=finalList, columns=["Company", "Industry","Location","Followers"])
linkRaw1000.to_csv('linkRaw1000.csv')
```
